```python
import cv2
from tensorflow.keras.models import load_model
import numpy as np

# Load the trained model
model = load_model('mask_detector2.keras')

# Initialize the camera
cap = cv2.VideoCapture(0)  # Use 0 for the default camera, or specify a different camera index

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame (resize, normalize, etc.)
    # Example: resize to match the input size of your model
    processed_frame = cv2.resize(frame, (224, 224))  # Assuming input size is 224x224

    # Make predictions
    # Example: assuming your model expects input in the shape (1, 224, 224, 3)
    prediction = model.predict(np.expand_dims(processed_frame, axis=0))

    # Display the frame and prediction
    cv2.putText(frame, 'Formal' if prediction[0][0] > 0.50 else 'Informal', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Frame', frame)

    # Exit on 'q' press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cap.release()
cv2.destroyAllWindows()
```
